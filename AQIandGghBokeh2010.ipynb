# This is my attempt to plot the AQI data in Bokeh. It would be AQI vs day of year in one county. 

```python
#import needed libraries
import numpy as np
import pandas as pd
from bokeh.plotting import *
from bokeh.models import ColumnDataSource
output_notebook()
output_file("index2010.html", title='States AQI through 2010')
```

```python
#read in data
data_all = pd.read_csv('daily_aqi_by_county_2010.csv')
df25 = data_all[(data_all['Defining Parameter'] == 'PM2.5')]
df25
```

```python
#convert date to day of year 
from dateutil.parser import parse
dateseries = df25['Date']
date_series = dateseries.map(lambda x: parse(x))
df25 = df25.assign(dayofyear = date_series.dt.dayofyear.tolist())
```

```python
import numpy as np

days = np.linspace(1,365,365)
alabama = np.zeros(365) - 1
```

```python
print(pd.unique(df25["State Name"]))
```

```python
statenames = ['Alaska', 'Washington' , 'Montana','North Dakota','Minnesota', 'Wisconsin','Michigan','New York','Vermont',
    'New Hampshire',  'Maine','Oregon','Idaho', 'Wyoming','South Dakota','Nebraska', 'Iowa','Illinois',
    'Indiana',   'Ohio','Pennsylvania', 'Connecticut', 'Rhode Island','Massachusetts', 
    'California', 'Nevada', 'Utah', 'Colorado', 'Kansas','Missouri', 'Kentucky', 'West Virginia', 'Virginia', 
    'District Of Columbia', 'Maryland','Delaware', 'New Jersey', 'Arizona', 'New Mexico','Texas','Oklahoma',
    'Arkansas', 'Tennessee', 'Louisiana', 'Mississippi','Alabama','Florida','Georgia',
    'South Carolina','North Carolina', 'Puerto Rico', 'Virgin Islands','Hawaii']
```

```python
newdf = pd.DataFrame()
newdf['DayofYear'] = days
for state in statenames:
    newdf[state] = np.zeros(365) -9999
    
```

```python
# convert the old data frame into a new one that arranges each value by state and day 

for i,day in enumerate(days):
    mydf = df25.loc[(df25['dayofyear'] == day)]
    for state in statenames:
        dfstate = mydf.loc[(mydf['State Name'] == state)]
        if len(dfstate) > 0:
            newdf[state].iloc[i]= np.mean(dfstate['AQI'])
newdf
```

```python
#saved new df to csv so I wouldn't have to calculate this again 
newdf.to_csv('meandayofyear2010.csv',index = False)
# newdf = pd.read_csv('meandayofyear.csv')
```

```python
source2 = ColumnDataSource(data = newdf)
#call tools for bokeh plot 
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select,lasso_select"


plotsarray = []
for state in statenames:
    p = figure(tools=TOOLS, width=350, height=350, title=state, y_range = (0,200))
    p.circle('DayofYear', state, source=source2)
    p.xaxis.axis_label = 'Day of Year'
    p.yaxis.axis_label = 'AQI for PM 2.5'
    plotsarray.append(p)
```

```python
show(gridplot(plotsarray,ncols=3))
```

# now I need to make a data frame that contains the average AQI category for each state at each day so I can make a df to use for the D3 map
#### good 0-50
#### moderate 51-100
#### unhealthy for sensitive groups 101-150
#### unhealthy 151-200
#### Very unhealthy 201-300
#### Hazardous 301-500

```python
catdf = pd.DataFrame()
for state in statenames:
    newcol = []
    for i in newdf[state]:
        if i > 0 and i <= 50 :
            newcol.append('Good')
        elif i > 50 and i <= 100:
            newcol.append('Moderate')
        elif i > 100 and i <= 150:
            newcol.append('Unhealthy for Sensitive Groups')
        elif i > 150 and i <= 200:
            newcol.append('Unhealthy')
        elif i > 200 and i <= 300:
            newcol.append('Very Unhealthy')
        elif i > 300 and i <= 500:
            newcol.append('Hazardous')
        else:
            newcol.append('null')
    catdf[state] = newcol
```

```python
catdf.to_csv('categoriesforStates2010.csv', index = False)
```
